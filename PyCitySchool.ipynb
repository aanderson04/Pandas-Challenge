```python
# dependecies and setup

import pandas as pd
import numpy as np
import csv

# load csv
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Create a high level snapshot (in table form) of the district's key metrics, including:

district_summary_df= pd.DataFrame(school_data_complete)
district_summary_df

# Total Schools
schools= district_summary_df["school_name"].unique()
total_schools= len(schools)
total_schools

# Total Students
total_students= district_summary_df["student_name"].count()
total_students

# Total Budget
budget= district_summary_df["budget"].unique()
total_budget= sum(budget)
total_budget

# Average Math Score
avg_math_score= district_summary_df["math_score"].mean()
avg_math_score

# Average Reading Score
avg_read_score= district_summary_df["reading_score"].mean()
avg_read_score

# %Passing Math (The percentage of students that passed math.)
passing_math= (district_summary_df[school_data_complete['math_score']>=70]['student_name'].count()/total_students)*100
passing_math

# %Passing Reading (The percentage of students that passed reading.)
passing_read= (district_summary_df[school_data_complete['reading_score']>=70]['student_name'].count()/total_students)*100
passing_read

# %Overall Passing (The percentage of students that passed math and reading.)
overall_pass= ((passing_math + passing_read)/2)
overall_pass
```

```python
# Create a dataframe to hold the above results

district= {
    'Total Schools':total_schools,
    'Total Students':'{:,}'.format(total_students),
    'Total Budget':'${:,.2f}'.format(total_budget),
    'Average Math Score':avg_math_score,
    'Average Reading Score':avg_read_score,
    '% Passing Math':passing_math,
    '% Passing Reading':passing_read,
    '% Overall Passing Score':[overall_pass]
}

district_table= pd.DataFrame(district)
district_table
```

```python
#Create an overview table that summarizes key metrics about each school, including:
# School Name
school_group = district_summary_df.groupby(['school_name'])
school_group

# School Type
school_type= school_group['type'].first()
school_type

# Total Students
students_per_school= school_group.size()
students_per_school

# Total School Budget
school_budget= school_group['budget'].first()
school_budget

# Per Student Budget
student_budget= (school_budget/students_per_school)
student_budget

# Average Math Score
school_avg_math= school_group['math_score'].mean()
school_avg_math

# Average Reading Score
school_avg_read= school_group['reading_score'].mean()
school_avg_read

# %Passing Math (The percentage of students that passed math.)
school_math_pass= district_summary_df[school_data_complete['math_score']>=70].groupby(['school_name']).size()
school_math_percent= (school_math_pass/ total_students)*100
school_math_percent

# %Passing Reading (The percentage of students that passed reading.)
school_read_pass= district_summary_df[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
school_read_percent= (school_read_pass/ total_students)*100
school_read_percent

#  %Overall Passing (The percentage of students that passed math and reading.)
school_overall_percent= (school_math_percent/ school_read_percent)/2
school_overall_percent
```

```python
#Create a dataframe to hold the above results

school_summary_dataframe = pd.DataFrame({"School Name":  school_group["school_name"],
                                "School Type": school_type,
                               "Total Students": students_per_school,
                               "Total School Budget": school_budget,
                               "Per Student Budget": student_budget, 
                               "Average Math Score":round(school_avg_math,2),
                               "Average Reading Score":round(school_avg_read,2), 
                               "% Passing Math": school_math_percent,
                               "% Passing Reading": school_read_percent,
                               "% Overall Passing Rate": school_overall_percent}) 
#format the new dataframe
school_summary_dataframe["Total Students"] = school_summary_dataframe["Total Students"].map("{:,.0f}".format)
school_summary_dataframe["Total School Budget"] = school_summary_dataframe["Total School Budget"].map("${:,.2f}".format)
school_summary_dataframe["Per Student Budget"] = school_summary_dataframe["Per Student Budget"].map("${:,.2f}".format)

school_summary_dataframe
```

```python
#Create a table that highlights the top 5 performing schools based on % Overall Passing. 
#Include: School Name, School Type, Total Students, Total School Budget, Per Student Budget, Average Math Score, Average Reading Score
#% Passing Math (The percentage of students that passed math.), % Passing Reading (The percentage of students that passed reading.), % Overall Passing (The percentage of students that passed math and reading.)

top_5_schools = school_summary_dataframe.sort_values(by='% Overall Passing Rate',ascending=False)
top_5_schools.head()
```
